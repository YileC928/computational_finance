## Exercise 18

```python
import numpy as np
sample_size = 100_000
rng = np.random.default_rng(seed=12345) 
u = rng.random(size=sample_size)
```

```python
import matplotlib.pyplot as plt
for mu in [1, 2, 3, 4]:
    samples = -np.log(u)/mu #because 1-u is also uniformly distributed
    samples_numpy = rng.exponential(size=sample_size)

    _, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 8))
    axs[0].hist(samples, bins=100, density=True);    # density argument changes the y-label. experiment!
    axs[0].set_title('Histogram of random numbers generated by inverse transform method, mu = {}'.format(mu))
    axs[1].hist(samples_numpy, bins=100, density=True);
    axs[1].set_title('Histogram of random numbers generated by NumPy directly, mu = {}'.format(mu));
    for ax in axs:                 # doing the same thing for both panels ('axes')
        ax.set_ylim([0, 1]);       # changing the limits on the y axis
        ax.set_xlim([0, 8]);
    plt.show();
```

## Exercise 21

```python
#1, 2, 3, 5, 8, 13, ... 
def fib(n):
    if n == 1:
        return 1
    elif n == 2:
        return 2
    else:
        return fib(n-2) + fib(n-1)
```

```python
fib(3)
```
